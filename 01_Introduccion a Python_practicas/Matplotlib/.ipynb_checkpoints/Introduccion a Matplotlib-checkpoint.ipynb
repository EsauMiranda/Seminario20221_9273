### Introduccion a Matplotlib

Matplotlib es una biblioteca de python que permite generar gr√°ficos a partir de datos que se encuentran en matrices o arrays

```python
# Importar las bibliotecas necesarios. En este caso numpy para realizar los calculos y trabajar con arreglos, matplotlib para realizar los graficos y pandas para importar datos y trabajar con DataFrames
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```python
# Crear una lista que vaya de 0 a 5 con un salto de 0.1 entre cada elemento
x_axis = np.arange(0, 5, 0.1)
x_axis
```

```python
# Crear una serie exponencial a partir de los valores contenidos en la variable x_axis
e_x = [np.exp(x) for x in x_axis]
e_x
```

```python
# Crear una gr√°fica a partir de los valores que tenemos en las variables x_axis y e_x
plt.plot(x_axis, e_x)
```

```python
# Asignar etiquetas a nuestros ejes
plt.xlabel("Time With MatPlotLib")
plt.ylabel("How Cool MatPlotLib Seems")

# mostrar el gr√°fico
plt.plot(x_axis, e_x)
plt.show()
```

Matplotlib tiene varias funciones que permiten modificar la presentacion de los graficos

```python
x_axis = np.arange(0, 10, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

```python
#Asignamos graficos para cada par de valores
#Cada punto del seno esta representado por un c√≠rculo rosa
sine_handle, = plt.plot(x_axis, sin, marker ='o', color='pink', label="Sine")
# Cada punto del coseno esta representado por un triangululo morado
cosine_handle, = plt.plot(x_axis, cos, marker='^', color='purple', label="Cosine")
#Definimos las etiquetas de los ejes
plt.xlabel("x")
plt.ylabel("F(x)")
#Asignamos el titulo del grafico
plt.title("Sin & Cos")
# Posicionamos las etiquetas de la gr√°fica
plt.legend(loc="upper right")
# Agregamos una red al gr√°fico
plt.grid()
#Definir los limites de los valores de la grafica
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

Existen diferentes tipos de graficos que pueden realizarse mediante matplitlib

#### Grafica de barras

```python
# Crear un arreglo con los valores que contendr√° el eje Y
users = [13000, 26000, 52000, 30000, 9000]
x_axis = np.arange(len(users))
```

```python
#Indicar a matplotlib el tipo de grafico que queremos realizar, en este caso, una grafica de barras
plt.bar(x_axis, users, color='r', alpha=0.5, align="center")
#Asignar las etiquetas para cada valor del eje x
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Java", "C++", "Python", "Ruby", "Clojure"])
# Designar los limites sobre el eje X
plt.xlim(-0.75, len(x_axis)-0.25)
# Designar los limites sobre el eje Y
plt.ylim(0, max(users)+5000)
# Dar un titulo y etiquetas sobre los ejes del gr√°fico
plt.title("Popularity of Programming Languages")
plt.xlabel("Programming Language")
plt.ylabel("Number of People Using Programming Languages")
```

#### Grafica de pastel

```python
#Definimos las etiquetas para las categor√≠as del gr√°fico
labels = ["Humans", "Smurfs", "Hobbits", "Ninjas"]

# Definimos los valores de cada secci√≥n
sizes = [220, 95, 80, 100]

#Asignamos los colores para cada secci√≥n
colors = ["red", "orange", "lightcoral", "lightskyblue"]

# Le decomps a matplotlib que "separe" cada secci√≥n
explode = (0.1, 0, 0, 0)

# Crear el grafico de pastel a partir de los valores definidos anteriormente
# Autom√°ticamente encuentra que porceentaje corresponde a cada secci√≥n
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
```

#### Scatteer plot

```python
import random
import pandas as pd
```

```python
x_limit = 100
#Definimos los valores del eje x
x_axis = np.arange(0,x_limit,1)
#Definimos los valores del eje y
data = [random.random() for value in x_axis]
datos = pd.DataFrame()
datos["Eje_x"] = x_axis
datos["Eje_y"] = data
```

```python
datos.plot(kind = 'scatter', x= 'Eje_x', y='Eje_y', grid = True, figsize = (8,8), title = 'Scatter plot')
```

#### Se pueden poner sub gr√°ficos dentro de un solo gr√°fico

```python
from scipy.stats import linregress
```

```python
# Set data
x_axis = np.arange(0, 10, 1)
fake = [1, 2.5, 2.75, 4.25, 5.5, 6, 7.25, 8, 8.75, 9.8]
```

```python
# Set line
(slope, intercept, _, _, _) = linregress(x_axis, fake)
fit = slope * x_axis + intercept
```

```python
# Plot data
fig, ax = plt.subplots()

fig.suptitle("Fake Banana Data!", fontsize=16, fontweight="bold")

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

ax.set_xlabel("Fake Banana Ages (in days)")
ax.set_ylabel("Fake Banana Weights (in Hundres of Kilograms)")

ax.plot(x_axis, fake, linewidth=0, marker='o')
ax.plot(x_axis, fit, 'r--')

plt.show()
```
